```python
import pandas as pd
from missforest.missforest import MissForest
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

IMF_original_merged = pd.read_csv("IMF_original_merged (1).csv") # read in data

# drop non-numerical values
countries = IMF_original_merged["Country"]
IMF_original_merged.drop(columns=["Country"], inplace=True)

rgr = RandomForestRegressor(n_jobs=-1) # base for the MF model is a random forest regressor; n_jobs=-1 means use all processors

mf = MissForest(rgr) # initialize MF model

filled_IMF = mf.fit_transform(IMF_original_merged)
filled_IMF["Country"] = countries # put country column back

# make a list of columns to keep
list_of_cols = ["Country"]
for year in range(1992, 2021):
  list_of_cols.append("T"+str(year))
```

```python
T_imputed = filled_IMF.loc[:, list_of_cols]
T_imputed
```

```python
from google.colab import files
T_imputed.to_csv('T_imputed.csv',index=False)
files.download('T_imputed.csv')
```
