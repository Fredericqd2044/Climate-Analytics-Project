```python
import matplotlib.pyplot as plt
```

Graph RMSE vs. MAE for applicable imputation methods

```python
Completely_Random_RMSE_FS1 = [0.965]
Completely_Random_MAE_FS1 = [0.790]
SI_by_Col_RMSE_FS1 = [.51, .517] # SimpleImputer by column -- mean and median
SI_by_Col_MAE_FS1 = [.39, .386]
ML_imputations_RMSE_FS1 = [.232, .24, .149, .154, .103, .061, .082]
ML_imputations_MAE_FS1 = [.173, .165, .026, .026, .016, .01, .014]
Our_RF_RMSE_FS1 = [0.387]
Our_RF_MAE_FS1 = [0.363]


SI_by_Col_RMSE_FS2 = [.491, .5] # SimpleImputer by column -- mean and median
SI_by_Col_MAE_FS2 = [.382, .377]
ML_imputations_RMSE_FS2 = [.385, .275, .064, .069, 1.772, .052, .048]
ML_imputations_MAE_FS2 = [.556, .419, .005, .005, .044, .004, .004]
Our_RF_RMSE_FS2 = [0.382]
Our_RF_MAE_FS2 = [0.357]
```

```python
# plot RMSE vs. MAE for FS1
plt.rcParams.update({'font.size': 10})
plt.scatter(Completely_Random_RMSE_FS1, Completely_Random_MAE_FS1, color="black", label="Random") # random
plt.scatter(SI_by_Col_RMSE_FS1, SI_by_Col_MAE_FS1, color="blue", label="Normal") # normal methods
plt.scatter(ML_imputations_RMSE_FS1, ML_imputations_MAE_FS1, color="red", label="ML") # ML imputations
plt.scatter(Our_RF_RMSE_FS1, Our_RF_MAE_FS1, color="red", marker="^") # Our RF model
plt.title("RMSE vs. MAE Temperature Feature Set")
plt.xlabel("RMSE")
plt.ylabel("MAE")
plt.legend()
plt.show()
```

```python
# plot RMSE vs. MAE for FS2
plt.scatter(Completely_Random_RMSE_FS1, Completely_Random_MAE_FS1, color="black", label="Random") # random
plt.scatter(SI_by_Col_RMSE_FS2, SI_by_Col_MAE_FS2, color="blue", label="Normal") # normal methods
plt.scatter(ML_imputations_RMSE_FS2, ML_imputations_MAE_FS2, color="red", label="ML") # ML imputations
plt.scatter(Our_RF_RMSE_FS2, Our_RF_MAE_FS2, color="red", marker="^") # Our RF model
plt.title("RMSE vs. MAE Extended Feature Set")
plt.xlabel("RMSE")
plt.ylabel("MAE")
plt.legend()
plt.show()
```

```python
RMSEs_FS1 = Completely_Random_RMSE_FS1 + SI_by_Col_RMSE_FS1 + ML_imputations_RMSE_FS1 + Our_RF_RMSE_FS1
MAEs_FS1 = Completely_Random_MAE_FS1 + SI_by_Col_MAE_FS1 + ML_imputations_MAE_FS1 + Our_RF_MAE_FS1

RMSEs_FS2 = SI_by_Col_RMSE_FS2 + ML_imputations_RMSE_FS2 + Our_RF_RMSE_FS2
MAEs_FS2 = SI_by_Col_MAE_FS2 + ML_imputations_MAE_FS2 + Our_RF_MAE_FS2
```

```python
plt.scatter(RMSEs_FS1, MAEs_FS1, color="sandybrown", label="Temperature FS")
plt.scatter(RMSEs_FS2, MAEs_FS2, color="limegreen", label="Extended FS")
plt.scatter(Completely_Random_RMSE_FS1, Completely_Random_MAE_FS1, color="black") # random
plt.title("RMSE vs. MAE Both Feature Sets")
plt.xlabel("RMSE")
plt.ylabel("MAE")
plt.legend(loc="upper right")
plt.show()
```

```python
ML_models = ["Iterative\nImputer\n", "KNN Imputer", "EM", "LOCF", "Moving\nWindow", "Iterative\nSVD", "MissForest", "Our RF\nModel"]
avg_RMSE_and_MAE_ML_models_FS1 = [.2025, .2025, .0875, .09, .0595, .0355, .048, 0.375]
avg_RMSE_and_MAE_ML_models_FS2 = [.4705, .347, .0345, .037, .908, .028, .026, 0.3695]
FS1_vs_FS2_ML_model_results_diffs = []
for i in range(len(avg_RMSE_and_MAE_ML_models_FS1)):
  FS1_vs_FS2_ML_model_results_diffs.append(avg_RMSE_and_MAE_ML_models_FS1[i] - avg_RMSE_and_MAE_ML_models_FS2[i])

plt.figure(figsize=(15,8))
plt.rcParams.update({'font.size': 14})
plt.bar(ML_models, FS1_vs_FS2_ML_model_results_diffs)
plt.title('Change in Average Metrics From Temperature to Extended Feature Set')
plt.xlabel('ML Imputation Model', fontsize=16)
plt.ylabel('Change in the Average of RMSE and MAE', fontsize=16)
plt.axhline(0, color='black')
plt.show()
```
