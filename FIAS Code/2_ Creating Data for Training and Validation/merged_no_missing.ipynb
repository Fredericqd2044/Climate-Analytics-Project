```python
import pandas as pd
import numpy as np

from google.colab import drive
from google.colab import files
drive.mount('/content/drive')
```

```python
# load all datasets
T_filepath = '/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/Individual Preprocessed Datasets/T_preprocessed.csv'
T_preprocessed = pd.read_csv(T_filepath)

C_filepath = '/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/Individual Preprocessed Datasets/C_preprocessed.csv'
C_preprocessed = pd.read_csv(C_filepath)

F_filepath = '/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/Individual Preprocessed Datasets/F_preprocessed.csv'
F_preprocessed = pd.read_csv(F_filepath)

L_filepath = '/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/Individual Preprocessed Datasets/L_preprocessed.csv'
L_preprocessed = pd.read_csv(L_filepath)

D_filepath = '/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/Individual Preprocessed Datasets/D_preprocessed.csv'
D_preprocessed = pd.read_csv(D_filepath)
```

```python
# merge datasets onto the temp dataset
merged_df = pd.merge(T_preprocessed, C_preprocessed, on='Country', how='left')
merged_df = pd.merge(merged_df, F_preprocessed, on='Country', how='left')
merged_df = pd.merge(merged_df, L_preprocessed, on='Country', how='left')
merged_df = pd.merge(merged_df, D_preprocessed, on='Country', how='left')
```
