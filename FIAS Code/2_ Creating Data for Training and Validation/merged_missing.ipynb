```python
import numpy as np
import pandas as pd

# this function just chooses random index positions to delete later
def binary_sampler(p, rows, cols): # got this code from: https://github.com/jsyoon0823/GAIN/blob/master/utils.py
  '''Sample binary random variables.

  Args:
    - p: probability of 1
    - rows: the number of rows
    - cols: the number of columns

  Returns:
    - binary_random_matrix: generated binary random matrix.
  '''
  np.random.seed(77)
  unif_random_matrix = np.random.uniform(0., 1., size = [rows, cols])
  binary_random_matrix = 1*(unif_random_matrix < p)
  binary_random_matrix[:, 29:] = 1
  return binary_random_matrix
```

```python
data = pd.read_csv('merged_no_missing.csv')
countries = data["Country"]
data = data.drop(columns=["Country"]) # need to include this only so it doesn't drop the countries

# Parameters
rows, cols = data.shape

miss_rate = 0.05 # what percent of data to remove -- similar to as in the original data

# Make missing data
data_m = binary_sampler(1-miss_rate, rows, cols) # choose what index positions to delete
merged_missing = data.copy()
merged_missing[data_m == 0] = np.nan # delete the index positions

# add back in the country column and put it as the frst column
merged_missing["Country"] = countries
cols = merged_missing.columns.tolist()
cols = cols[-1:] + cols[:-1]
merged_missing = merged_missing[cols]

print(merged_missing) # this is the dataset with some deleted
```
