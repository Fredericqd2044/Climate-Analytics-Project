```python
# import methods
import pandas as pd
```

```python
# original data
T = pd.read_csv('Annual_Surface_Temperature_Change.csv')
T
```

```python
# Keep Country and Years 1992-2020 columns
selected_columns = ['Country'] + [f'F{year}' for year in range(1992, 2021)]
T_selected = T[selected_columns]
T_selected
```

```python
# Rename years columns to T[Year]
renamed_columns = {'Country': 'Country'}
for year in range(1992, 2021):
    renamed_columns[f'F{year}'] = f'T{year}'

T_selected = T_selected.rename(columns=renamed_columns)
T_selected
```

```python
# extract to csv file
file_path = "T_preprocessed.csv"
T_selected.to_csv(file_path, index=False)
print(f"T_selected has been saved to {file_path}")
```
