```python
import pandas as pd
from google.colab import drive
from google.colab import files
drive.mount('/content/drive')

path = '/content/drive/MyDrive/IMF_Climate-Related_Disasters.csv'
df = pd.read_csv(path)
df.head()
```

```python
df_new = df[df['Indicator'].str.contains("TOTAL", case=False)]
df_new_filled = df_new.fillna(0)
columns = ["Country"] + [f"F{year}" for year in range(1992, 2021)]
df_final = df_new_filled[columns]
df_final.columns = ["Country"] + [f"D{year}" for year in range(1992, 2021)]

df_final.head()
```

```python
df_final.to_csv('Climate_Disaster.csv', index=False)
files.download('Climate_Disaster.csv')
```
