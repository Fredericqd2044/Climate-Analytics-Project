```python
file_path = '/content/drive/My Drive/Land_Cover_Accounts.csv'

# Load the CSV data into a DataFrame
df = pd.read_csv(file_path)
df.head(10)
```

```python
#Check different indicators in the dataframe
print(df['Indicator'].unique())
print()

#Check columns name
print(df.columns)
df.head()
```

```python
#Select rows where indicator is "Climate Altering Land Cover Index"
filtered_df = df[df['Indicator'] == 'Climate Altering Land Cover Index']

#removed columns we don't need
filtered_df.drop(['ObjectId','ISO2','ISO3','Indicator', 'Unit', 'Source','CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor', 'Climate_Influence'], axis=1, inplace=True)
```

```python
#check the filtered_data
filtered_df.head(5)
```

```python
#check country name in the dataframe
country_name=filtered_df['Country'].values
print(country_name)

#There are country names like'advanced Economies','Africa', 'Americas'...
#Remove these country names
index=int(np.where(country_name == 'World')[0])+1
filtered_df=filtered_df[index:]
```

```python
#check the filtered_df after removed certain territory
filtered_df.columns = [col.replace('F', 'L') for col in filtered_df.columns]
filtered_df
```

```python
#check info of filtered_df
filtered_df.info()

# Check for null values in each row
null_rows = filtered_df.isnull().any(axis=1)

# Display rows with at least one null value
filtered_df[null_rows]
```

```python
#Convert dataframe into csv file
filtered_df.to_csv('Land_Cover_Index.csv', index=False)
files.download('Land_Cover_Index.csv')
```
