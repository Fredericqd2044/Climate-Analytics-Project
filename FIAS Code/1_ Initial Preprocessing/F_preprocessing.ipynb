```python
# import methods
import pandas as pd
```

```python
# original data
F = pd.read_csv('Forest_and_Carbon.csv')
F
```

```python
# Isolate Forest Area rows and only the countries
forest_area = F[F['Indicator'] == 'Forest area']
forest_area
```

```python
# Make a dataset where the columns are Country, F1992, F1993…F2020
selected_columns = ['Country'] + [f'F{year}' for year in range(1992, 2021)]
forest_area = forest_area[selected_columns]
forest_area
```

```python
# extract to csv file
file_path = "F_preprocessed.csv"
forest_area.to_csv(file_path, index=False)
print(f"forest_area has been saved to {file_path}")
```
