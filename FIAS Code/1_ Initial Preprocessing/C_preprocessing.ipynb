```python
import pandas as pd
import numpy as np

from google.colab import drive
from google.colab import files
drive.mount('/content/drive')

CO2_data_filepath = '/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/IMF_Atmospheric_CO2.csv'
CO2_IMF_data = pd.read_csv(CO2_data_filepath)
CO2_IMF_data.head(5)
```

```python
CO2_IMF_data = CO2_IMF_data[["Date", "Value"]]

years = []

for date in CO2_IMF_data["Date"]:
  year = int(date[:4])
  years.append(year)

CO2_IMF_data["Year"] = years
CO2_IMF_data.drop(columns=["Date"], inplace=True)
CO2_IMF_data.head(5)
```

```python
ST_data_filepath = '/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/IMF_Surface_Temperature.csv'
ST_IMF_data = pd.read_csv(ST_data_filepath)
countries = list(ST_IMF_data["Country"])
```

```python
year_avg_CO2s = []
year_col_names = []
for year in CO2_IMF_data.Year.unique():
  year_col_names.append("C"+str(year))
  year_data = CO2_IMF_data[CO2_IMF_data["Year"] == year]
  year_avg_CO2s.append(sum(year_data["Value"]) / len(year_data["Value"]))

year_avg_CO2_rows_expanded = np.tile(year_avg_CO2s[34:-4], (len(countries),1))

CO2_data_cleaned = pd.DataFrame(data = year_avg_CO2_rows_expanded, columns=year_col_names[34:-4])
CO2_data_cleaned["Country"] = countries
CO2_data_cleaned
```

```python
CO2_data_cleaned.to_csv('CO2.csv',index=False)
files.download('CO2.csv')
```
