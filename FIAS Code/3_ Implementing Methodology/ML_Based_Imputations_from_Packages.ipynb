```python
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error
def RMSE(data_real, data_imputed):
  RMSE = np.sqrt(mean_squared_error(data_real, data_imputed))
  return round(RMSE, 3)
def MAE(data_real, data_imputed):
  return round(mean_absolute_error(data_real, data_imputed), 3)
```

```python
from google.colab import drive
from google.colab import files
drive.mount('/content/drive')
np.random.seed(77)

T_missing = pd.read_csv('/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/T_missing.csv')
T_no_missing = pd.read_csv('/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/T_no_missing.csv')
merged_missing = pd.read_csv('/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/merged_missing.csv')
merged_no_missing = pd.read_csv('/content/drive/MyDrive/Spring 2024/Advanced Topics in Data Science/MLCC Project/data/merged_no_missing.csv')

countries = T_no_missing["Country"]
T_missing.drop(columns=["Country"], inplace=True)
T_no_missing.drop(columns=["Country"], inplace=True)
merged_missing.drop(columns=["Country"], inplace=True)
merged_no_missing.drop(columns=["Country"], inplace=True)

T_missing_np = np.array(T_missing)
T_no_missing_np = np.array(T_no_missing)
merged_missing_np = np.array(merged_missing)
merged_no_missing_np = np.array(merged_no_missing)
```

```python
import impyute

EM_FS1 = impyute.imputation.cs.em(T_missing_np)
print("EM Feature Set 1 RMSE:", RMSE(T_no_missing_np, EM_FS1))
print("EM Feature Set 1 MAE:", MAE(T_no_missing_np, EM_FS1))
print()
EM_FS2 = impyute.imputation.cs.em(merged_missing_np)
print("EM Feature Set 2 RMSE:", RMSE(merged_no_missing_np, EM_FS2))
print("EM Feature Set 2 MAE:", MAE(merged_no_missing_np, EM_FS2))
```

```python
LOCF_FS1 = impyute.imputation.ts.locf(T_missing_np, axis=1) # axis is 1 because each column is a different time
print("LOCF Feature Set 1 RMSE:", RMSE(T_no_missing_np, LOCF_FS1))
print("LOCF Feature Set 1 MAE:", MAE(T_no_missing_np, LOCF_FS1))
print()
LOCF_FS2 = impyute.imputation.ts.locf(merged_missing_np, axis=1) # axis is 1 because each column is a different time
print("LOCF Feature Set 2 RMSE:", RMSE(merged_no_missing_np, LOCF_FS2))
print("LOCF Feature Set 2 MAE:", MAE(merged_no_missing_np, LOCF_FS2))
```

```python
MW_FS1 = impyute.imputation.ts.moving_window(T_missing_np) # axis is 1 because each column is a different time
print("Moving Window Feature Set 1 RMSE:", RMSE(T_no_missing_np, MW_FS1))
print("Moving Window Feature Set 1 MAE:", MAE(T_no_missing_np, MW_FS1))
print()
MW_FS2 = impyute.imputation.ts.moving_window(merged_missing_np) # axis is 1 because each column is a different time
print("Moving Window Feature Set 2 RMSE:", RMSE(merged_no_missing_np, MW_FS2))
print("Moving Window Feature Set 2 MAE:", MAE(merged_no_missing_np, MW_FS2))
```

```python
from fancyimpute import IterativeSVD

ISVD_FS1 = IterativeSVD().fit_transform(T_missing_np)
print("Iterative SVD Feature Set 1 RMSE:", RMSE(T_no_missing_np, ISVD_FS1))
print("Iterative SVD Feature Set 1 MAE:", MAE(T_no_missing_np, ISVD_FS1))
print()
ISVD_FS2 = IterativeSVD().fit_transform(merged_missing_np)
print("Iterative SVD Feature Set 2 RMSE:", RMSE(merged_no_missing_np, ISVD_FS2))
print("Iterative SVD Feature Set 2 MAE:", MAE(merged_no_missing_np, ISVD_FS2))
```

```python
from missforest.missforest import MissForest
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

rgr = RandomForestRegressor(n_jobs=-1) # base for the MF model is a random forest regressor; n_jobs=-1 means use all processors

mf = MissForest(rgr) # initialize MF modem

MF_FS1 = mf.fit_transform(T_missing) # use the original pandas dfs for this; plug in missing data, get out imputed data
print("MissForest Feature Set 1 RMSE:", RMSE(T_no_missing, MF_FS1))
print("MissForest Feature Set 1 MAE:", MAE(T_no_missing, MF_FS1))
```

```python
MF_FS2 = mf.fit_transform(merged_missing)
print("MissForest Feature Set 2 RMSE:", RMSE(merged_no_missing, MF_FS2))
print("MissForest Feature Set 2 MAE:", MAE(merged_no_missing, MF_FS2))
```
