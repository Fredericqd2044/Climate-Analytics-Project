```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error
import random
random.seed(77)

file_path = '/content/drive/MyDrive/T_no_missing.csv'

# Load the CSV data into a DataFrame
data = pd.read_csv(file_path)
data.head(10)

#remove country for later calculation
data = data.drop(columns=["Country"])
data
```

```python
miss_data = pd.read_csv('/content/drive/MyDrive/T_missing.csv')
miss_data_x = miss_data.drop(columns=["Country"])
miss_data_x
```

```python
def random_impute_uniform(series):
    min_val = series.min()
    max_val = series.max()
    return series.apply(lambda x: random.uniform(min_val, max_val) if pd.isna(x) else x)

imputed_data_x_df = miss_data_x.copy()
for column in miss_data_x.columns:
    imputed_data_x_df[column] = random_impute_uniform(miss_data_x[column])

imputed_data_x_df
```

# Calculate the MSE, RMSE, MAE of the Random Imputation Method

```python
data=data.to_numpy()
imputed_data_x=imputed_data_x_df.to_numpy()
mask = np.isnan(miss_data_x)
true_values = data[mask]
imputed_values = imputed_data_x[mask]

#calculate the MSE
mse = mean_squared_error(true_values, imputed_values)
print(f"Mean Squred Error (MSE) of random imputation: {mse}")

#calculate the RMSE
rmse = np.sqrt(mean_squared_error(true_values, imputed_values))
print(f"Root Mean Squred Error (RMSE) of random imputation: {rmse}")

# Calculate MAE only for the imputed values
mae = mean_absolute_error(true_values, imputed_values)
print(f"Mean Absolute Error (MAE) of random imputation: {mae}")
```
