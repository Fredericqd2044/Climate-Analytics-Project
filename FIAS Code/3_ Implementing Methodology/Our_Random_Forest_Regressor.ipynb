# Baseline Model






```python
#get temperature avaliable year for each country
temp_list_all=[]

for c in range(len(df_missing)):
  temp=df_missing.iloc[c][1:30]
  temp_missing=temp[temp.isnull()].index
  temp_list=temp.drop(temp_missing).index
  temp_list_all.append(temp_list)
```

```python
def generate_baseline_features(temp_list_all, df_missing, index):
    # Define the columns for the DataFrame
    df_columns = ['Year', 'CTemp_Year_min1', 'CTemp_Year_min2', 'CTemp_Year_min3', 'Temp_Change_Year']
    features = pd.DataFrame(columns=df_columns)

    # Select the temperature list and row based on the given index
    temp_list = temp_list_all[index]
    row = df_missing.iloc[index]

    # Loop over each temperature entry in the list
    for t in temp_list:
        Year = int(t.replace('T', ''))
        extended_value = []

        if Year >= 1995:
            # Process for years >= 1995
            for i in range(3):
                past_year = "T" + str(Year - (i + 1))
                year_value = row[past_year] if past_year in row else np.nan

                # Handle missing values by checking if the value is NaN and backfilling
                if np.isnan(year_value) and (Year - (i + 2)) >= int(temp_list[0].replace('T', '')):
                    past_year = "T" + str(Year - (i + 2))
                    year_value = row[past_year] if past_year in row else np.nan

                    if np.isnan(year_value):
                        two_years_back = "T" + str(Year - (i + 3))
                        year_value = row[two_years_back] if two_years_back in row else np.nan

                extended_value.append(year_value)

        else:
            # Process for years < 1995
            for i in range(3):
                past_year = "T" + str(Year - (i + 1))
                year_value = row[past_year] if past_year in row else 9999999

                extended_value.append(year_value)

        # Create a new row for each year and append to the DataFrame
        new_row = pd.DataFrame([{
            'Year': str(Year),
            'CTemp_Year_min1': extended_value[0],
            'CTemp_Year_min2': extended_value[1],
            'CTemp_Year_min3': extended_value[2],
            'Temp_Change_Year': row[t]
        }])
        features = pd.concat([features, new_row], ignore_index=True)

    # Year started from 1995
    features = features[3:]
    return features
```

```python
# Store results for analysis
results = []

# Loop through each index from 0 to 125
for idx in range(126):
  features=generate_baseline_features(temp_list_all,df_missing, index=idx)
  X=features.drop(['Year','Temp_Change_Year'],axis=1)
  y=features['Temp_Change_Year']

  #train and test split
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=77)

  # Initialize the Random Forest Regressor
  rf_regressor_baseline = RandomForestRegressor(n_estimators=100, random_state=77)

  # Initialize the StandardScaler
  scaler = StandardScaler()

  # Fit the scaler on the training and test data and transform it
  X_train_scaled = scaler.fit_transform(X_train)
  X_test_scaled = scaler.transform(X_test)

  # Train the model
  rf_regressor_baseline.fit(X_train_scaled, y_train)

  # Make predictions on the test set
  predictions = rf_regressor_baseline.predict(X_test_scaled)

  # Calculate the mean squared error
  mse = mean_squared_error(y_test, predictions)
  mae = mean_absolute_error(y_test, predictions)

  results.append({
        'Index': idx,
        'Country':df_missing['Country'].iloc[idx],
        'MAE': mae,
        'MSE': mse
    })
```

# Extend Feature Set

```python
def generate_extended_features(temp_list_all, df_missing, index):
    # Initialize a DataFrame to store the transformed data
    extended_features_columns = ['Year','C_Year-1','F_Year-1','L_Year-1','D_Year-1']
    extended_features = pd.DataFrame(columns=extended_features_columns)

    # Select the temperature list and row based on the given index
    temp_list = temp_list_all[index]
    row = df_missing.iloc[index]

    #find out corresponding values -CO2 concentration, Forest for t-1 year...
    for t in temp_list:
      Year=int(t.replace('T',''))
      Past_Year=Year-1

      #tempereature related variables-last year data
      C_column= 'C'+str(Past_Year)
      F_column= 'F'+str(Past_Year)
      L_column= 'L'+str(Past_Year)
      D_column= 'D'+str(Past_Year)

      temp_related_list=[C_column,F_column,L_column,D_column]

      if (Year>1992):
        temp_related_values=row[temp_related_list]

        # Prepare the new row to add to the DataFrame
        new_row = pd.DataFrame([{
              'Year': str(Year),
              'C_Year-1': temp_related_values[0],
              'F_Year-1': temp_related_values[1],
              'L_Year-1': temp_related_values[2],
              'D_Year-1': temp_related_values[3],
          }])
      else:
        new_row = pd.DataFrame([{
              'Year': str(Year),
              'C_Year-1': 9999999,
              'F_Year-1': 9999999,
              'L_Year-1': 9999999,
              'D_Year-1': 9999999,
          }])


      # Append the new row to the DataFrame
      extended_features = pd.concat([extended_features, new_row], ignore_index=True)

    #print out the dataframe for train and validate
    extended_features=extended_features[3:]
    return extended_features
```

```python
# Store results for analysis
results_extended_features = []

# Loop through each index from 0 to 125
for idx in range(126):
  baseline_features=generate_baseline_features(temp_list_all, df_missing, index=idx)
  extended_features=generate_extended_features(temp_list_all, df_missing, index=idx)
  merged_df=merge_extended_features(baseline_features, extended_features)

  X=merged_df.drop(['Year','Temp_Change_Year'],axis=1)
  y=merged_df['Temp_Change_Year']

  #train and test split
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=77)

  # Initialize the Random Forest Regressor
  rf_regressor_extended = RandomForestRegressor(n_estimators=100, random_state=77)

  # Initialize the StandardScaler
  scaler = StandardScaler()

  # Fit the scaler on the training and test data and transform it
  X_train_scaled = scaler.fit_transform(X_train)
  X_test_scaled = scaler.transform(X_test)

  # Train the model
  rf_regressor_extended.fit(X_train_scaled, y_train)

  # Make predictions on the test set
  predictions = rf_regressor_extended.predict(X_test_scaled)

  # Calculate the mean squared error
  mse = mean_squared_error(y_test, predictions)
  mae = mean_absolute_error(y_test, predictions)

  results_extended_features.append({
        'Index': idx,
        'Country':df_missing['Country'].iloc[idx],
        'MAE': mae,
        'RMSE': np.sqrt(mse)
    })
```
